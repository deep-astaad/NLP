```python
from nltk import word_tokenize
import tkinter as tk
from tkinter import *

def preprocess(d):
    d=d.lower()
    d="eos "+ d
    d=d.replace("."," eos")
    return d

def generate_tokens(d):
    tokens = word_tokenize(d)
    return tokens

def generate_tokens_freq(tokens):
    frq={}
    for i in tokens:
        frq[i]=0
    for i in tokens:
        frq[i]+=1
    return frq

def generate_ngrams(tokens,k):
    l=[]
    i=0
    while(i<len(tokens)):
        l.append(tokens[i:i+k])
        i=i+1
    l=l[:-1]
    return l

def generate_ngram_freq(bigram):
    frq1={}
    for i in bigram:
        st=" ".join(i)
        frq1[st]=0
    for i in bigram:
        st=" ".join(i)
        frq1[st]+=1
    return frq1

def find1(s,frq1):
    try:
        return frq1[s]
    except:
        return 0

def print_probability_table(distinct_tokens,frq,frq1):
    n=len(distinct_tokens)
    l=[[]*n for i in range(n)]
    for i in range(n):
        denominator = frq[distinct_tokens[i]]
        for j in range(n):
            numerator = find1(distinct_tokens[i]+" "+distinct_tokens[j],frq1)
            l[i].append(float("{:.3f}".format(numerator/denominator)))
    return l

def result():
    d = inputtxt.get(1.0, "end-1c")

    d=preprocess(d)

    tokens=generate_tokens(d)

    distinct_tokens = list(set(sorted(tokens)))
    frq=generate_tokens_freq(tokens)

    bigram = generate_ngrams(tokens,2)

    for i in bigram:
        print("'{}'".format(' '.join(i)), end=", ")

    frq1=generate_ngram_freq(bigram)
    

    probability_table = print_probability_table(distinct_tokens,frq,frq1)

    printOut.insert(END, "Probability Table :")
    printOut.insert(END, '\n')
    n=len(distinct_tokens)
    
    printOut.insert(END, '\t')
    for i in range(n):
        printOut.insert(END, distinct_tokens[i] + "\t")
    printOut.insert(END, "\n")
    for i in range(n):
        printOut.insert(END, distinct_tokens[i])
        printOut.insert(END, '\t')
        for j in range(n):
            printOut.insert(END, probability_table[i][j])
            printOut.insert(END, '\t')
        printOut.insert(END, "\n")
    
    probability(probability_table,distinct_tokens)
    
def probability(probability_table,distinct_tokens):
    text = check_prob.get(1.0, "end-1c")
    
    p = preprocess(text)

    t=generate_tokens(p)
    frqs=generate_tokens_freq(t)
    n = generate_ngrams(t,2)
    s=1
    frq2={}
    
    for i in n:
        frq2[" ".join(i)]=0
    for i in n:
        k=distinct_tokens.index(i[0])
        m=distinct_tokens.index(i[1])
        frq2[" ".join(i)]=probability_table[k][m]
        s*=probability_table[k][m]
    printOut.insert(END, "\n")
    printOut.insert(END, "Calculate Probability of the sentence: ")
    x=frq2.popitem()
    printOut.insert(END, "{:.5f}".format(s))

frame = tk.Tk()
frame.title("TextBox Input")
frame.geometry('1200x1200')

inputtxt = tk.Text(frame,
                   height = 3,
                   width = 50)
inputtxt.pack()
check_prob = tk.Text(frame,
                   height = 3,
                   width = 50)
check_prob.pack()

print_prob = tk.Button(frame,
                        text = "Print Probability", 
                        command = result)
print_prob.pack()

printOut = tk.Text(height = 37, width = 150, background = 'grey', font = 'calibri')
printOut.pack()
frame.mainloop()

```
