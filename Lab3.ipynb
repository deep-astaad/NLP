```python
# Suppoted forms
# NOUN      -> base, singular, plural, proper singular, proper plural
# ADJECTIVE -> base, comparative, comparative adjective, superlative, superlative adjective
# VERB      -> base, past, gerund, past participlenon, 3rd person singular present, 3rd person singular present
# ADVERB    -> base, comparative, superlative


# nlp code
import tkinter as tk
from tkinter import *
import spacy
import pyinflect
nlp = spacy.load("en_core_web_sm")


# noun
def function_noun(token):
    root = token.text
    singular = token._.inflect("NN")
    plural = token._.inflect("NNS")
    proper_singular = token._.inflect("NNP")
    proper_plural = token._.inflect("NNPS")

    # input noun forms
    if root[:1] == singular[:1]:
        output_window.insert(END, "singular noun: " + singular)
        output_window.insert(END, '\n')
    if root[:1] == plural[:1]:
        output_window.insert(END, "plural noun: " + plural)
        output_window.insert(END, '\n')
    # elif form == 'Proper Singular':
    #     output_window.insert(END, "proper noun, singular: " + proper_singular)
    #     output_window.insert(END, '\n')
    # elif form == 'Proper Plural':
    #     output_window.insert(END, "proper noun, plural: " + proper_plural)
    #     output_window.insert(END, '\n')

# adjective
def function_adjective(token):
    root = token.text
    base = token._.inflect("JJ")
    comparative = token._.inflect("JJR")
    superlative = token._.inflect("JJS")
    # if root[:-1] == 'y':
    get1 = base+"ly"
    check1 = nlp(get1)
    get2 = base+"ness"
    check2 = nlp(get2)
    
    
    # get3 = base_adjective+"ily"
    # input adjectives forms
    # for token1 in check1:
    #     if token1.lemma_ == root:
    #         output_window.insert(END, "ly form: " + get1)
    #         output_window.insert(END, '\n')
    if root[:1] == base[:1]:
        output_window.insert(END, "base form: " + base)
        output_window.insert(END, '\n')
    if root[:1] == comparative[:1]:
        output_window.insert(END, "comparative adjective: " + comparative)
        output_window.insert(END, '\n')
    if root[:1] == superlative[:1]:
        output_window.insert(
            END, "superlative adjective: " + superlative)
        output_window.insert(END, '\n')
    
    # output_window.insert(END, "ly form: " + get1)
    # output_window.insert(END, '\n')
    # output_window.insert(END, "ness form: " + get2)
    # output_window.insert(END, '\n')
    # if :
    #     output_window.insert(
    #         END, "superlative adjective: " + superlative_adjective)
    #     output_window.insert(END, '\n')
        
# verb
def function_verb(token):
    root = token.text
    base = token._.inflect("VB")
    past = token._.inflect("VBD")
    present_participle = token._.inflect("VBG")
    past_participle = token._.inflect("VBN")
    non_3rd_singular_present = token._.inflect("VBP")
    third_singular_present = token._.inflect("VBZ")

    # input tense forms
    if root[:1] == base[:1]:
        output_window.insert(END, "base form: " + base)
        output_window.insert(END, '\n')
    if root[:1] == past[:1]:
        output_window.insert(END, "past tense: " + past)
        output_window.insert(END, '\n')
    if root[:1] == present_participle[:1]:
        output_window.insert(END, "present participle: " + present_participle)
        output_window.insert(END, '\n')
    if root[:1] == past_participle[:1]:
        output_window.insert(END, "past participle: " + past_participle)
        output_window.insert(END, '\n')
    if root[:1] == third_singular_present[:1]:
        output_window.insert(
            END, "3rd person singular present: " + third_singular_present)
        output_window.insert(END, '\n')
    
    

# adverb
def function_adverb(token):
    root = token.text
    base = token._.inflect("RB")
    comparative = token._.inflect("RBR")
    superlative = token._.inflect("RBS")

    # input adverbs forms
    if root[:1] == base[:1]:
        output_window.insert(END, "base form: " + base)
        output_window.insert(END, '\n')
    if root[:1] == comparative[:1]:
        output_window.insert(END, "comparative adverb: " + comparative)
        output_window.insert(END, '\n')
    if root[:1] == superlative[:1]:
        output_window.insert(END, "superlative adverb: " + superlative)
        output_window.insert(END, '\n')

# input from the gui window
def printResults():
    user_input = input_word.get(1.0, "end-1c")
    doc = nlp(user_input)
    # selection = tense.get()
    # print(selection)
    # print(selection)
    for token in doc :
        # checking parts of speech tags
        # var = input_form.get(1.0, "end-1c")
        # var = "NULL"
        output_window.insert(END, "pos: " + token.pos_)
        output_window.insert(END, '\n')
        output_window.insert(END, "tag: " + token.tag_)
        output_window.insert(END, '\n')
        # noun
        if token.pos_ == 'NOUN':
            function_noun(token)

        # adjective
        elif token.pos_ == 'ADJ':
            function_adjective(token)

        # verb
        elif token.pos_ == 'VERB':
            # selection=tense.get()
            # print(selection)
            function_verb(token)

        # adverb
        elif token.pos_ == 'ADV':
            function_adverb(token)
        else :
            output_window.insert(END, "Please Enter NOUN, VERB, ADVERB and ADJECTIVE")
            output_window.insert(END, '\n')
    output_window.insert(END, '----------------------------')
    output_window.insert(END, '\n')

# def clearSCR() :
#     output_window.delete('1.0', END)

# gui code
gui = tk.Tk()
gui.title('Different Morphological Forms')
gui.config(background='light grey')
gui.geometry('450x250')

# Input window
input_word = tk.Text(gui, height=1, width=20, bg="light blue")
input_word.pack()
# print_word = tk.Button(gui, text="Input Word", command=take_input)
# print_word.pack()


printButton = tk.Button(gui,
                        text = "Result", 
                        command = printResults)
printButton.pack()

# Input form
# input_form = tk.Text(gui, height=1, width=20, bg="light pink")
# input_form.pack()
# print_word = tk.Button(gui, text="Print Given Form", command=take_input)
# print_word.pack()


#Create a dropdown Menu
# tense= StringVar()
# tense.set("Select Tense")
# tense= OptionMenu(gui, tense,"Base", "Past","Past Participle","Gerund", command=take_input)
# tense.pack()
# print_word = tk.Button(gui, text="Print Given Form", command=take_input)
# print_word.pack()

# gender= StringVar()
# gender.set("Select Gender")
# gender= OptionMenu(gui, gender,"Male", "Female", command=take_input)
# gender.pack()

# number= StringVar()
# number.set("Select Number")
# number= OptionMenu(gui, number,"Singular", "Plural", command=take_input)
# number.pack()

# degree= StringVar()
# degree.set("Select Degree")
# degree= OptionMenu(gui, degree,"Normal", "Comparative", "Superlative", command=take_input)
# degree.pack()

#Create a dropdown Menu
# tense= OptionMenu(gui, tense,"Base", "Past","Past Participle","Gerund")
# tense.pack()

# ouput window
output_label = tk.Label(gui, text="", height=1, width=40, font=("black"))
output_label.pack()
output_window = tk.Text(gui, height=500, width=40, bg="light cyan")
output_window.pack()

# output_window.delete('1.0', END)
# clear_scr = tk.Button(gui, text="Clear Screen", command=clearSCR)
# clear_scr.pack()

# input_word.insert(0,"Enter Word")
# input_form.insert(0, "Enter Form")
# execute the guiect
gui.mainloop()
```
